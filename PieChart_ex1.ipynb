```python
import matplotlib.pyplot as plt

# Dados de exemplo (tipos de alimentos e quantidades)
emocoes = ["Tristeza", "Alegria", "Medo", "Desgosto", "Raiva"]
quantidades = [0.6203125, 0.1425, 0.41197916666666667, 0.41015625, 0.32083333333333336]

# Plotar o gráfico de pizza
plt.pie(quantidades, labels=emocoes, autopct='%1.1f%%', startangle=140)

# Personalização do gráfico
plt.title('Distribuição de Emoções na Análise de Sentimentos')

# Mostrar o gráfico
plt.savefig ("grafiv.png")

```
